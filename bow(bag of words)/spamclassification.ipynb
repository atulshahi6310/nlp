```python
import pandas as pd

# Sample dataset
data = {
    "message": [
        "Congratulations! Congratulations!  You won a free iPhone. Click here.",
        "Hey, are we still meeting for lunch today?",
        "Limited time offer! Congratulations!  Claim your prize now.",
        "Don't forget to submit the assignment.",
        "Win cash now!!! Visit this link.",
        "Can you call me when you get home?"
    ],
    "label": [
        "spam",
        "ham",
        "spam",
        "ham",
        "spam",
        "ham"
    ]
}

# Create DataFrame
df = pd.DataFrame(data)
```

```python
import nltk
import re
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
courpus =[]
for i in range(0,len(df)):
    review = re.sub('[^a-zA-Z]',' ',df['message'][i])
    review=review.lower()
    review=review.split()
    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    review =' '.join(review)
    courpus.append(review)
```

```python
courpus
```

## bow,tfidf,ml algo for the project 

## Bow

```python
from sklearn.feature_extraction.text import CountVectorizer
count = CountVectorizer(max_features=100 , binary=True,ngram_range=(1,3)) ## pick the top 2500 words have max number of frequency
```

```python
x =count.fit_transform(courpus).toarray()
```

```python
x ## independent feature
```

```python
count.vocabulary_
```

```python
y = pd.get_dummies(df['label']).astype(int)
```

```python
y
```

```python
y = y.iloc[:,0].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.20)
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
spam_detect_model = MultinomialNB().fit(x_train,y_train)
```

```python
y_pred = spam_detect_model.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import classification_report,accuracy_score
```

```python
report = classification_report(y_pred,y_test)
```

```python
print(report)
```
