```python
import pandas as pd

# Sample dataset
data = {
    "message": [
        "Congratulations! Congratulations!  You won a free iPhone. Click here.",
        "Hey, are we still meeting for lunch today?",
        "Limited time offer! Congratulations!  Claim your prize now.",
        "Don't forget to submit the assignment.",
        "Win cash now!!! Visit this link.",
        "Can you call me when you get home?"
    ],
    "label": [
        "spam",
        "ham",
        "spam",
        "ham",
        "spam",
        "ham"
    ]
}

# Create DataFrame
df = pd.DataFrame(data)
```

```python
df
```

```python
import nltk
import re
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
courpus =[]
for i in range(0,len(df)):
    review = re.sub('[^a-zA-Z]',' ',df['message'][i])
    review=review.lower()
    review=review.split()
    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    review =' '.join(review)
    courpus.append(review)
```

```python
courpus
```

## create the bag of word

```python
from sklearn.feature_extraction.text import CountVectorizer
count = CountVectorizer(max_features=2500 , binary=True) ## pick the top 2500 words have max number of frequency
```

```python
x =count.fit_transform(courpus).toarray()
```
