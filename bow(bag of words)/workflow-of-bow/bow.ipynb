# Bag of word

```python
corpus = '''Bag of Words (BoW) is a text representation technique used in Natural Language Processing (NLP) to convert text into numerical vectors so that machine learning algorithms can understand and process it.
The main idea of BoW is:
A document is represented by the words it contains and their frequencies, ignoring grammar and word order.
It is called “bag” of words because words are treated like items in a bag where order does not matter.'''
```

```python
# tokenization
from nltk.tokenize import word_tokenize
tokens = word_tokenize(corpus)
```

```python
#lemmatization
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
stopwords =stopwords.words("english")
lem = WordNetLemmatizer()
for words in tokens:
    tokens = [lem.lemmatize(words) for words in tokens if words not in stopwords]
    content =[ ' '.join(tokens)]
```

```python
print(content)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
x = CountVectorizer(max_features=2500 , binary=False,ngram_range=(1,3))
```

```python
count = x.fit_transform(content).toarray()
```
