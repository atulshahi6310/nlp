```python
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()
```

```python
import re
import nltk
nltk.download('stopwords')
```

```python
corpus = []
for i in range(0,len(data_set)):
    messages = re.sub('[^a-zA-Z]',' ' ,data_set['v2'][i])
    messages = messages.lower()
    messages = messages.split()

    messages =[lemmatizer.lemmatize(word) for word in messages]
    messages = ' '.join(messages)
    corpus.append(messages)
    
    
```

```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess
```

## The gensim.utils.simple_preprocess() function is a basic, convenient utility for converting a raw text document (string) into a list of cleaned, individual tokens (words). It performs several common preprocessing steps automatically

```python
word = []
for sent in corpus:
    sent_token = sent_tokenize(sent)
    for sent in sent_token:
        word.append(simple_preprocess(sent))
```
