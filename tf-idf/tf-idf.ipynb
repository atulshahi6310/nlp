```python
import pandas as pd

data = {
    "message": [
        "Congratulations! Congratulations! You won a free iPhone. Click here.",
        "Hey, are we still meeting for lunch today?",
        "Limited time offer! Congratulations! Claim your prize now.",
        "Don't forget to submit the assignment.",
        "Win cash now!!! Visit this link.",
        "Hey, are we still meeting for lunch today?",
        "Limited time offer! Congratulations! Claim your prize now.",
        "Don't forget to submit the assignment.",
        "Win cash now!!! Visit this link.",
        "Hey, are we still meeting for lunch today?",
        "Limited time offer! Congratulations! Claim your prize now.",
        "Don't forget to submit the assignment.",
        "Win cash now!!! Visit this link.",
        "Hey, are we still meeting for lunch today?",
        "Limited time offer! Congratulations! Claim your prize now.",
        "Don't forget to submit the assignment",
        "Win cash now!!! Visit this link.",
        "Can you call me when you get home?"
    ],
    
    "label": [
        "spam","ham","spam","ham","spam",
        "ham","spam","ham","spam",
        "ham","spam","ham","spam",
        "ham","spam","ham","spam","ham"
    ]
}

df = pd.DataFrame(data)

print(df.head())
```

```python
df.head()
```

```python
import nltk
import re
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
courpus =[]
for i in range(0,len(df)):
    review = re.sub('[^a-zA-Z]',' ',df['message'][i])
    review=review.lower()
    review=review.split()
    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    review =' '.join(review)
    courpus.append(review)
```

```python
courpus
```

## this is how tf-idf is implement

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf =TfidfVectorizer(
                        max_features=10
                        
                        )
```

```python
tfidf.fit_transform(courpus).toarray()
```

## apply the concepts of n-grams

```python
tfidf1 =TfidfVectorizer(
                        max_features=30,
                        ngram_range=(1,2)
                        
                        )
```

```python
tfidf1.fit_transform(courpus).toarray()
```
