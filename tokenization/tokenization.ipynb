## 📘 Theory of Tokenization (NLP)

**Tokenization** is the process of breaking raw text into smaller meaningful units called **tokens**.
Tokens can be:

* **Sentences**
* **Words**
* **Subwords**
* **Characters**

It is the **first step in Natural Language Processing (NLP)** before tasks like POS tagging, parsing, sentiment analysis, or transformers.

---

### 🔹 Why Tokenization is Important?

1. Converts raw text into structured format
2. Helps in feature extraction (Bag of Words, TF-IDF)
3. Required before feeding text into ML/DL models
4. Handles punctuation, contractions, symbols properly

---

### 🔹 Types of Tokenization

1. **Sentence Tokenization** → Splitting text into sentences
2. **Word Tokenization** → Splitting sentences into words
3. **Subword Tokenization** → Used in Transformers (BPE, WordPiece)
4. **Character Tokenization** → Splitting into characters

In traditional NLP (like using NLTK), we mainly use **sentence and word tokenization**.

---

# 📌 NLTK Tokenizers

Import:

```python
import nltk

```

---

# 1️⃣ `sent_tokenize()`

### 🔹 Definition

`sent_tokenize()` splits a paragraph into individual sentences.

### 🔹 How it Works

It uses **Punkt Sentence Tokenizer**, which:

* Learns abbreviation patterns
* Detects sentence boundaries using punctuation like `. ! ?`
* Avoids breaking at abbreviations (e.g., Dr., Mr., etc.)

### 🔹 Example

```python
text = "Hello Atul. You are learning NLP! That's great."
sent_tokenize(text)
```

### 🔹 Output

```python
['Hello Atul.', 'You are learning NLP!', "That's great."]
```

### 🔹 Use Case

* Chatbot responses
* Text summarization
* Document segmentation

---

# 2️⃣ `word_tokenize()`

### 🔹 Definition

`word_tokenize()` splits text into words and punctuation.

### 🔹 How it Works

It:

* Uses Punkt sentence tokenizer internally
* Splits words properly
* Keeps punctuation as separate tokens
* Handles contractions (e.g., "don't" → "do", "n't")

### 🔹 Example

```python
text = "I don't like NLP!"
word_tokenize(text)
```

### 🔹 Output

```python
['I', 'do', "n't", 'like', 'NLP', '!']
```

### 🔹 Key Feature

✔️ Linguistically smart
✔️ Handles contractions properly

---

# 3️⃣ `wordpunct_tokenize()`

### 🔹 Definition

Splits text into words and punctuation using **regular expressions**.

### 🔹 How it Works

* Splits at whitespace
* Separates all punctuation
* More aggressive splitting

### 🔹 Example

```python
text = "I don't like NLP!"
wordpunct_tokenize(text)
```

### 🔹 Output

```python
['I', 'don', "'", 't', 'like', 'NLP', '!']
```

### 🔹 Difference from `word_tokenize()`

| word_tokenize     | wordpunct_tokenize  |
| ----------------- | ------------------- |
| Smart splitting   | Regex based         |
| "don't" → do, n't | "don't" → don, ', t |
| More accurate     | More basic          |

---

# 4️⃣ `TreebankWordTokenizer`

### 🔹 Definition

Tokenizer based on **Penn Treebank** rules.

📚 The **Penn Treebank** is a large annotated corpus of English text used in NLP research.

### 🔹 How it Works

* Follows linguistic rules
* Splits contractions properly
* Separates punctuation carefully
* Keeps decimal numbers intact

### 🔹 Example

```python
tokenizer = TreebankWordTokenizer()
text = "I don't like NLP. It costs $5.50."
tokenizer.tokenize(text)
```

### 🔹 Output

```python
['I', 'do', "n't", 'like', 'NLP.', 'It', 'costs', '$', '5.50', '.']
```

### 🔹 Special Features

✔️ Handles contractions like word_tokenize
✔️ Keeps decimal numbers intact
✔️ Slightly more rule-based

---

# 🔥 Comparison Summary

| Feature              | sent_tokenize | word_tokenize | wordpunct_tokenize | TreebankWordTokenizer |
| -------------------- | ------------- | ------------- | ------------------ | --------------------- |
| Splits Sentences     | ✅             | ❌             | ❌                  | ❌                     |
| Splits Words         | ❌             | ✅             | ✅                  | ✅                     |
| Handles Contractions | ❌             | ✅             | ❌                  | ✅                     |
| Regex Based          | ❌             | ❌             | ✅                  | ❌                     |
| Linguistically Smart | Medium        | High          | Low                | High                  |

---

# 🎯 When to Use What?

* 📄 Document splitting → `sent_tokenize`
* 🤖 NLP pipeline → `word_tokenize`
* ⚡ Fast basic splitting → `wordpunct_tokenize`
* 📚 Linguistic research → `TreebankWordTokenizer`
