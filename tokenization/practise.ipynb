```python
courpus = '''naik's.
Natural Language Processing (NLP) is fun!
I love learning about AI, machine learning, and data science.
ChatGPT helps people write code, emails, and stories.
Tokenization splits text into words, subwords, or sentences.
In 2026, many students study NLP online. 
'''
```

```python
## tokenization
## paragraph ----> sentencess
from nltk.tokenize import sent_tokenize
```

```python
sent_tokenize(courpus)
```

## word_tokenize

```python
## tokenization 
##paraghargh --- > words
## sentences ---> words
from nltk.tokenize import word_tokenize
```

```python
words = word_tokenize(courpus)
```

```python
print(words)
len(words)
```
