# Stemming

```python
corpus ='''Pichai Sundararajan (born June 10, 1972), better known as Sundar Pichai (pronounced: /ˈsʊndɜːr pɪˈtʃeɪ/),
is an Indian–American business executive who has been the CEO of Google since 2015 and the CEO of its parent company Alphabet Inc.
since 2019.
Pichai began his career as a materials engineer. Following a short stint at the management consulting firm McKinsey & Co., Pichai joined Google in 2004,[4] where he led the product management and innovation efforts for a suite of Google's client software products, including Google Chrome and ChromeOS, as well as being largely responsible for Google Drive. In addition, he went on to oversee the development of other applications such as Gmail and Google Maps.
As of February 2026, his net worth is estimated at US$1.6 billion
'''
```

```python
print(corpus)
```

# Word-tokenize

```python
from nltk.tokenize import word_tokenize
tokens = word_tokenize(corpus)
```

```python
print(tokens)
```

# Stemming --porter-steemmer

```python
from nltk.stem import PorterStemmer
stemming = PorterStemmer()
```

```python
for word in tokens:
    print(word + "----->"+stemming.stem(word))
```

# Regexpstemmer

```python
from nltk.stem import RegexpStemmer
stemming = RegexpStemmer('ing$|s$|e$|abc$') ## We can define regex patterns (using OR '|') inside RegexpStemmer
## to remove specific suffixes from words 
for words in tokens:
    print(words+"--------->"+stemming.stem(words))
```

# Snowball stemmer --- porter2stemmer

```python
from nltk.stem import SnowballStemmer
stem = SnowballStemmer('english')
for words in tokens:
    words = words.lower()
    print(words+ "------->"+stem.stem(words))
```
