```python
from nltk.tokenize import word_tokenize
tokens = word_tokenize(corpus)
```

```python
# print(tokens)
# print(len(tokens))
```

```python
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
stopwords = stopwords.words("english")
lem = PorterStemmer()
tokens = [lem.stem(words) for words in tokens if words not in stopwords]
```

```python
# print(tokens)#
# print(len(tokens))
```

```python
from nltk.tag import pos_tag
poss = pos_tag(tokens) 
```

```python
# print(poss)
```

```python
from nltk.chunk import ne_chunk
name_entity = ne_chunk(poss).draw() ## this gives us graph  which is use to detech the name  or  noune 
```
